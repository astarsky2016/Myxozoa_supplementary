```python
DNMT = set()
TET = set()
BER = set()
enzymes = [BER, DNMT, TET]
files = ["BER.txt", "DNMT.txt", "TET.txt"]
for pos, f in enumerate(files):
    with open(f, "r") as f_in:
        for line in f_in:
            data = line.split()
            if len(data) >= 2:
                enzymes[pos].add(data[0])
test = {"DNMT":DNMT, "TET":TET, "BER": BER}
for k in test.keys():
    print (k, len(test[k]))
```

```python
from collections import defaultdict
import os
#dodaj host informaciju u tablicu
counter = defaultdict(set)
import glob
taxa = [os.path.basename(f).replace(".fna", "") for f in glob.glob("/disk3/ALL_ASSEMBLIES/ORIGINAL_ASSEMBLY/*.fna")]
predictions = glob.glob("/disk3/ALL_ASSEMBLIES/ORIGINAL_ASSEMBLY/*per_pred.tsv")
panthers = glob.glob("/disk3/ALL_ASSEMBLIES/ORIGINAL_ASSEMBLY/*_panther.out")
counter = defaultdict(set)
pth_counter = defaultdict(set)
myxo = {}
for org in taxa:
    tax_info = {}
    tax_data = defaultdict(dict)
    for pred in predictions:
        if org in pred:
            with open(pred, "r") as tax_in:
                for line in tax_in:
                    data = line.split("\t")
                    tax_info[data[0]] = data[2:4]
    for pthr in panthers:
        if org in pthr:
            with open(pthr, "r") as pthr_in:
                if org == "sample_58":
                    print(pthr)
                for line in pthr_in:
                    data = line.split("\t")
                    tax_hit = tax_info.get(data[0], None)
                    pt_id = data[1]
                    pth_counter[pt_id].add(org)
                    for k, v in test.items():
                        if pt_id in v:
                            counter[k].add(pt_id)
                            tax_data[k][pt_id] = tax_hit
    myxo[org] = tax_data
            
        
```

```python
data = {}
samples = []
pathways = []
families = []
taxons = []
species = []
counts = []
for org in myxo.keys():
    for enz in myxo[org].keys():
        for path in myxo[org][enz].keys():
            samples.append(org)
            pathways.append(enz)
            families.append(path)
            taxons.append(myxo[org][enz][path][0])
            species.append(myxo[org][enz][path][1])
            counts.append(len(pth_counter[path]))
data["Sampled species"] = samples
data["Methylation machinery"] = pathways
data["Panther protein family"] = families
data["Assigned taxon"] = taxons
data["Assigned species"] = species
data["Presence"] = counts
for k, v in data.items():
    print(k, len(v))
print (len(data["Presence"]))
counter = defaultdict(int)
for m in data["Methylation machinery"]:
    counter[m] += 1
print (counter)
```
