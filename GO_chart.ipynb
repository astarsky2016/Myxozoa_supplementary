```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pickle
```

```python
data = pickle.load(open("graph_lists_70.p", "rb"))
func = ["Biological Process", "Molecular Function", "Cellular Component"]
```

```python
# Extract function names (X-axis labels) and percentages (Y-axis values)
percentages1 = [item[1] for item in data[func[0]]]
percentages2 = [item[1] for item in data[func[1]]]
percentages3 = [item[1] for item in data[func[2]]]
# Sort the data based on percentages in descending order
sorted_data1 = sorted(zip(percentages1, functions1), reverse=True)
sorted_data2 = sorted(zip(percentages2, functions2), reverse=True)
sorted_data3 = sorted(zip(percentages3, functions3), reverse=True)
# Create an array of indices for the X-axis locations
#functions1 = [item[0] for item in data[func[0]]]
#functions2 = [item[0] for item in data[func[1]]]
#functions3 = [item[0] for item in data[func[2]]]
#x1 = np.arange(len(functions1))
#x2 = np.arange(len(functions2))
#x3 = np.arange(len(functions3))
# Select the top 20 values and labels
top_values1 = [val for val, _ in sorted_data1[:20]]
top_labels1 = [label for _, label in sorted_data1[:20]]
top_values2 = [val for val, _ in sorted_data2[:20]]
top_labels2 = [label for _, label in sorted_data2[:20]]
top_values3 = [val for val, _ in sorted_data3[:20]]
top_labels3 = [label for _, label in sorted_data3[:20]]
```

```python
# Set the width of the bars
bar_width = 0.3

# Create a 1x3 subplot layout for the three organisms
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
spec = fig.add_gridspec(1, 3, width_ratios=[2, 2, 2])


# Create bars for each organism and add them to respective subplots
bar1 = axs[0].bar(np.arange(len(top_values1)), top_values1, bar_width, label=func[0], color='blue')
bar2 = axs[1].bar(np.arange(len(top_values2)), top_values2, bar_width, label=func[1], color='red')
bar3 = axs[2].bar(np.arange(len(top_values3)), top_values3, bar_width, label=func[2], color='green')

# Set X-axis labels for each subplot
x_labels = [top_labels1, top_labels2, top_labels3]
for i, ax in enumerate(axs):
    ax.set_xticks(np.arange(len(top_labels1)))  # Use np.arange(len()) for top_values1, top_values2, top_values3
    ax.set_xticklabels(x_labels[i], rotation=45, ha="right", fontsize=8)

# Set Y-axis label for the first subplot
axs[0].set_ylabel('Percentage')
# Set Y-axis limits to range from 0 to 100%
for ax in axs:
    ax.set_ylim(0, 100)
# Add a legend to each subplot
for ax in axs:
    ax.legend()

# Add a title to each subplot
axs[0].set_title(func[0])
axs[1].set_title(func[1])
axs[2].set_title(func[2])

# Add a common title to the entire figure
fig.suptitle('Top 20 Biological Class Percentages by Organism', fontsize=16)

# Adjust spacing between subplots
#plt.tight_layout(rect=[0, 0, 1, 0.95])

# Show the plot
plt.show()
```
